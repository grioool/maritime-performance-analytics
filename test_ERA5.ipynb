```python
import cdsapi

URL = "https://cds.climate.copernicus.eu/api"
KEY = "KEY"

dataset = "reanalysis-era5-single-levels"
request = {
    "product_type": ["reanalysis"],
    "variable": [
        "10m_u_component_of_wind",
        "10m_v_component_of_wind",
        "significant_height_of_combined_wind_waves_and_swell"
    ],
    "year": ["2023"],
    "month": ["02"],
    "day": ["02"],
    "time": ["12:00"],
    "format": "netcdf",
    "data_format": "netcdf",
    "download_format": "unarchived",
    "area": [61.23, -2.12, 51.04, 11.2]
}

client = cdsapi.Client(url = URL, key = KEY)
client.retrieve(dataset, request).download()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt


ds_wind = xr.open_dataset("wind.nc", engine="h5netcdf")
ds_waves = xr.open_dataset("waves.nc", engine="h5netcdf")

print("Wind variables:", list(ds_wind.keys()))
print("Wave variables:", list(ds_waves.keys()))

# calculating wind speed from u and v components (vectors)
wind_speed = np.sqrt(ds_wind['u10']**2 + ds_wind['v10']**2)
ds_wind['wind_speed'] = wind_speed

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))


ds_waves['swh'].plot(ax=ax1, cmap='Blues')
ax1.set_title('Wave height (H_s) [m]')
ax1.grid(True, linestyle='--', alpha=0.5)


ds_wind['wind_speed'].plot(ax=ax2, cmap='YlOrRd')
ax2.set_title('Wind speed [m/s]')
ax2.grid(True, linestyle='--', alpha=0.5)

plt.suptitle('Exemplary data for North Sea', fontsize=16)

plt.show()
```
